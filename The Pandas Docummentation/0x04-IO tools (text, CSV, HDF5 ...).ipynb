### Orient options
 - There are different options for the format of the resulting JSON file / string.

```python
dfjo = pd.DataFrame(
    dict(A=range(1, 4), B=range(4, 7), C=range(7, 10)),
    columns=list("ABC"),
    index=list("xyz")
)
dfjo
```

```python
sjo = pd.Series(dict(x=15, y=16, z=17), name="D")

sjo
```

```python
#column oriented, default for dataframe serializes the data as nesed json objects.
dfjo.to_json(orient="columns")
```

```python
# index oriented the default for series similar to column oriented but the index labels are now primary.
dfjo.to_json(orient="index")
```

```python
sjo.to_json(orient="index")
```

```python
#record oriented serializes the dara to a JSON array of column -> value records index labels are not included and
#this is useful when the dta is needed for plotting libraries like the javascript d3.js
dfjo.to_json(orient="records")
```

```python
sjo.to_json(orient="records")
```

```python
#in value oriented the column and index labels are not included and only the bare-bones options which serializes to nested JSON.

dfjo.to_json(orient="values")
```

```python
#split oriented serializes to a JSON object containing separate entries for values, index nand columns.

dfjo.to_json(orient="split")
```

```python
sjo.to_json(orient="split")
```

### Normalization
 - pandas provides a utility function to take a dict or list of dicts and normalize this semi-structured data into a flat table.

```python
data = [
    {"id": 1, "name": {"first": "Coleen", "last": "Volk"}},
    {"name": {"given": "Mark", "family": "Regner"}},
    {"id": 2, "name": "Faye Raker"},
]

pd.json_normalize(data)
```

```python
data = [
    {
        "state": "Florida",
        "shortname": "FL",
        "info": {"governor": "Rick Scott"},
        "county": [
            {"name": "Dade", "population": 12345},
            {"name": "Broward", "population": 40000},
            {"name": "Palm Beach", "population": 60000},
        ],
    },
    {
        "state": "Ohio",
        "shortname": "OH",
        "info": {"governor": "John Kasich"},
        "county": [
            {"name": "Summit", "population": 1234},
            {"name": "Cuyahoga", "population": 1337},
        ],
    },
]

pd.json_normalize(data, "county", ["state", "shortname", ["info", "governor"]])
```

### Table Schema
- This is a spec for describing tabular datasets as a JSON object. The JSON includes information on the field names, types, and other attributes. You can use the orient `table` to build a JSON string with two fields `schema` and `data`.

```python
df = pd.DataFrame(
    {
        "A": [1, 2, 3],
        "B": ["a", "b", "c"],
        "C": pd.date_range("2016-01-01", freq="d", periods=3),
    },
    index=pd.Index(range(3), name="idx"),
)
df
```

```python
df.to_json(orient="table", date_format="iso")
```

```python
from pandas .io.json import build_table_schema

s = pd.Series(pd.date_range("2016", periods=4))

build_table_schema(s)
```

```python
# A `primarykey` field containing an array of labels, is included if the index is unique.

s_dupe = pd.Series([1, 2], index=[1, 1])

build_table_schema(s_dupe)
```

```python
# with multiple indexes the primary key wil be an array:
s_multi = pd.Series(1, index=pd.MultiIndex.from_product([("a", "b"), (0, 1)]))

build_table_schema(s_multi)
```

```python
# `read_json` accepts `orient="table"` as an arguement, allowing for the preservation of metadata such ad dtypes and index names
# in a round-trippable manner.

df = pd.DataFrame(
    {
        "foo": [1, 2, 3, 4],
        "bar": ["a", "b", "c", "d"],
        "baz": pd.date_range("2018-01-01", freq="d", periods=4),
        "qux": pd.Categorical(["a", "b", "c", "c"]),
    },
    index = pd.Index(range(4), name="idx")
)

df
```

```python
df.to_json("test.json", orient="table")

new_df = pd.read_json("test.json", orient="table")

new_df
```

### Reading HTML content

```python
#Reading a url without any options:

url = "https://www.fdic.gov/resources/resolutions/bank-failures/failed-bank-list"
pd.read_html(url)
```

```python
# Reading a URL while passing headers alongside the HTTP  request:

url = 'https://www.sump.org/notes/request/' # HTTP request reflector
pd.read_html(url)
```

```python
html_str = """
         <table>
             <tr>
                 <th>A</th>
                 <th colspan="1">B</th>
                 <th rowspan="1">C</th>
             </tr>
             <tr>
                 <td>a</td>
                 <td>b</td>
                 <td>c</td>
             </tr>
         </table>
     """


with open("tmp.html", "w") as f:
    f.write(html_str)


df = pd.read_html("tmp.html")

df[0]
```

```python
from io import StringIO

dfs = pd.read_html(StringIO(html_str))

dfs[0]
```

```python
# Read a URL and match a table that contains specific text:

dfs = pd.read_html(url, header=0)

#specifying an index column:
dfs = pd.read_html(url, index_col=0)

#specify a number of rows to skip:
dfs = pd.read_html(url, skiprows=0)

#specify a number of rows to skip using a list (`range` works well)
dfs = pd.read_html(url, skiprows=range(2))

#specify a HTML attribute.
dfs1 = pd.read_html(url, attrs={"id": "table"})
dfs2 = pd.read_html(url, attrs={"class": "sortable"})
print(np.array_equal(dfs1[0], dfs2[0])) #should be true

#specifying the values that should be converted to NaN
dfs = pd.read_html(url, na_values=["No Acquirer"])

#specify whether to keep the default set of NaN  values:
dfs = pd.read_html(url, keep_default_na=False)
```

```python
# converting numericak columns to strings to avoid the loss of the leading zeros.

url_mcc = "https://en.wikipedia.org/wiki/Mobile_country_code"

dfs = pd.read_html(
    url_mcc,
    match = "Telekom Albania",
    header=0,
    converters = {"MNC": str},
)
dfs
```

```python
dfs = pd.read_html(url, match="Metcalf Bank", index_col=0)
```
